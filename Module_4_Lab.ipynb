## Aggregation

```python
# aggregate the order_products data set to get the count of the number of items ordered per department 
# reset the index and rename the columns after you've aggregated the data, so the columns make sense for what is being displayed (i.e. the column for "count" should be named accordingly)
items = order_products[['department','product_id']].groupby(['department']).agg('count')
items = items.reset_index()
items.columns = ["Department","Number of Items Ordered"]
items
```

```python
# aggregate the order_products data set to get the count of the number of items ordered per aisle
# reset the index and rename the columns after you've aggregated the data, so the columns make sense for what is being displayed (i.e. the column for "count" should be named accordingly)
items = order_products[['aisle','product_id']].groupby(['aisle']).agg('count')
items = items.reset_index()
items.columns = ["Aisle","Number of Items Ordered"]
items
```

```python
# aggregate the order_products data set to get the count of the number of items ordered per day of week
# reset the index and rename the columns after you've aggregated the data, so the columns make sense for what is being displayed (i.e. the column for "count" should be named accordingly)

# To interpret the data, remember that 0 = Monday, 1 = Tuesday, 2 = Wednesday, etc.
items = order_products[['order_dow','product_id']].groupby(['order_dow']).agg('count')
items = items.reset_index()
items.columns = ["Day of the Week","Number of Items Ordered"]
items
```

```python
# aggregate the orders data set to get the mean number of items (num_items) ordered per order number (order_number)
# reset the index and rename the columns after you've aggregated the data, so the columns make sense for what is being displayed (i.e. the column for "count" should be named accordingly)
items = orders[['order_number','num_items']].groupby(['order_number']).agg('mean')
items = items.reset_index()
items.columns = ["Order Number","Mean of Items Ordered"]
items
```

```python
# aggregate the orders data set to get the median number of items (num_items) ordered per order number (order_number)
# reset the index and rename the columns after you've aggregated the data, so the columns make sense for what is being displayed (i.e. the column for "count" should be named accordingly)
items = orders[['order_number','num_items']].groupby(['order_number']).agg('median')
items = items.reset_index()
items.columns = ["Order Number","Median of Items Ordered"]
items
```

# Part B: Subsetting/Merging

```python
# Run this cell to create the example data sets needed for the next section
fruit_weights = pd.DataFrame({'fruit':['apples','bananas','apricots'],'weight_lbs':[0.50,.33,.10]})
fruit_prices = pd.DataFrame({'fruit':['oranges','bananas','apples'],'price_lbs':[1.39,0.58,2.99]}) 
```

## Subsetting

```python
# In fruit_weights, select rows where the fruit weight in pounds (weight_lbs) is less than .4 lbs (using the < operator)
weights = fruit_weights.loc[fruit_weights['weight_lbs']< 0.4]
weights
```

```python
# In fruit_prices, select rows where the fruit price per pound (price_lbs) is higher than $1.00/lb (using the > operator)
prices = fruit_prices.loc[fruit_prices['price_lbs']> 1]
prices
```

```python
# In fruit_prices, select only rows where fruit = 'apples'
apples = fruit_prices.loc[fruit_prices['fruit'] == 'apples']
apples
```

## Merging

```python
# Merge fruit_weights and fruit_prices, resulting in a dataframe called merge1. Your final data set should have *all* fruits that appear in each data set. 
# Before doing this, ask yourself "What type of join results in all records being pulled from both datasets?"
# Get the number of rows and print the first few rows
merge1 = pd.merge(fruit_weights,fruit_prices,how='outer',on=['fruit'])
merge1
```

```python
# Merge fruit_weights and fruit_prices, resulting in a dataframe called merge2.
# Your final data set should *only* have fruits for which there is a record in both the fruit_weights AND fruit_prices datasets
# Get the number of rows and print the first few rows
merge2 = pd.merge(fruit_weights,fruit_prices,how='inner',on=['fruit'])
merge2
```

```python
# Merge the organic and conventional data sets, resulting in a dataframe called merge3, using how = "inner", on aisle_id & aisle
# Get the number of rows and print the first few rows
merge3 = pd.merge(organic,conventional,how='inner',on=['aisle_id','aisle'])
print(merge3.shape[0])
merge3.head()
```

```python
# Merge the organic and conventional data sets, resulting in a dataframe called merge4, using how = "outer", on aisle_id & aisle
# get the number of rows and print the first few rows
merge4 = pd.merge(organic,conventional,how='outer',on=['aisle_id','aisle'])
print(merge4.shape[0])
merge4.head()
```

```python
# Merge the organic and conventional data sets, resulting in a dataframe called merge5, using how = "left", on aisle_id & aisle
# Get the number of rows and print the first few rows
merge5 = pd.merge(organic,conventional,how='left',on=['aisle_id','aisle'])
print(merge5.shape[0])
merge5.head()
```

```python
# Merge the organic and conventional data sets, resulting in a dataframe called merge6, with how = "right" , on aisle_id & aisle
# Get the number of rows and print the first few rows
merge6 = pd.merge(organic,conventional,how='right',on=['aisle_id','aisle'])
print(merge6.shape[0])
merge5.head()
```

# Part C: Reshaping/Calculating

## Reshaping

```python
# Run this cell to create the example data sets needed for the next section
fruit_orders = pd.DataFrame({'fruit':['apples','bananas','apricots'],'order1':[4,0,5],'order2':[1,3,0],'order3':[6,5,4]})
fruit_inventory = pd.DataFrame({'fruit':['apples','apples','oranges','oranges'],'store':['safeway','giant','safeway','giant'],'inventory':[100,75,62,59]})

```

```python
# convert fruit_orders to long format
fruit_orders = pd.melt(fruit_orders, id_vars='fruit', 
                      value_vars=['order1', 'order2', 'order3'], 
                      var_name='Order', value_name='Num of Orders')
fruit_orders
```

```python
# convert fruit_inventory to wide format
fruit_inventory = fruit_inventory.pivot(index='fruit', columns='store', values='inventory')
fruit_inventory
```

```python
# convert the merged conventional/organic data set with how="inner", merge3, to long format
merge3 = pd.melt(merge3, id_vars='aisle', 
                      value_vars=['num_items_x', 'aisle_id'] ,
                      var_name='category', value_name='value')
merge3
```

## Calculating

```python
# Create a new column in merge1, the fruit prices dataset, that equals the total price per item of each fruit
# HINT: price_per_item = price per pound (price_lbs) * lbs (weight_lbs)
# print the first few rows

merge1['price_per_item'] = merge1['price_lbs'] *merge1['weight_lbs']
merge1.head()
```
